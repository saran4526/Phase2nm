```python
new_transaction = {'Transaction ID': 1016, 'Date': '2025-01-18', 'Items Purchased': 'Milk, Bread'}
df = pd.concat([df, pd.DataFrame([new_transaction])], ignore_index=True)
df.tail()
```

```python
from collections import Counter
import matplotlib.pyplot as plt

items = df['Items Purchased'].str.cat(sep=', ').split(', ')
item_counts = Counter(items)

pd.Series(item_counts).sort_values(ascending=False).plot(kind='bar')
plt.title('Item Frequency')
plt.ylabel('Count')
plt.show()
```
